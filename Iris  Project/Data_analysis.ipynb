### Explanation:
* `Count`: The total number of observations or entries.
* `Mean`: The average value.
* `Std`: The standard deviation, which measures the amount of variation or dispersion in the values.
* `Min`: The smallest value.
* `25%`: The first quartile, or the value below which 25% of the observations fall.
* `Median`: The middle value when the data is sorted in ascending order. Also known as the 50th percentile.
* `75%`: The third quartile, or the value below which 75% of the observations fall.
* `Max`: The largest value.

#### 5. Data Cleaning

* Identifying and handling missing values (if necessary).
* Removing duplicates (if necessary).
* Converting data types (if necessary).

```python
# check for null values
data.isnull().sum()
```

```python
# check for the total amount of null values
print('the total amount of null values is:', data.isnull().sum().sum())
```

### Some techniques to deal with null values:
* `Drop Null Values`: You can use the `dropna()` function to remove rows or columns with null values.

* `Fill Null Values`: The `fillna()` function allows you to replace null values with a specific value or a computed value (like mean, median, or mode).
* `Interpolation`: The `interpolate()` function can estimate the missing values based on other values in the dataset. This is useful for time series data.

```python
# Check for duplicates
data[data.duplicated(keep= False) == True]
```

```python
# The total number of duplicates
print('The number of duplicates is:', data.duplicated().sum())
```

```python
# Remove duplicates
data.drop_duplicates(inplace= True) 
```

```python
# Review The total number of duplicates
print('The number of duplicates is:', data.duplicated().sum())
```

#### 6. Data Visualization

* Using Matplotlib or Seaborn for basic visualizations (scatter plots, bar charts, histograms).
* Explaining how to customize plots and add labels.
* Introducing plotly for advanced interactive visualizations.

```python
# plot sepal length vs width (without labels and legend)

plt.scatter(data['sepal_length'], data['sepal_width'])
```

```python
# plot sepal length vs width (witht labels)

plt.scatter(data['sepal_length'], data['sepal_width'])
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)'),
plt.title('Sepal Length vs Width')
```

```python
# plot sepal length vs width (witht labels and legend [colored by species])

# Create a scatter plot with a hue
sns.scatterplot(data=data, x="sepal_length", y="sepal_width", hue="species")

# Display the plot
plt.legend(loc='upper left')
plt.show()
```

```python
# plot petal length vs width (witht labels and legend [colored by species])

# Create a scatter plot with a hue
sns.scatterplot(data=data, x="petal_length", y="petal_width", hue="species")

# Display the plot
plt.legend(loc='upper left')
plt.show()
```

```python
# Add interactivity using plotly

import plotly.express as px

# Assuming 'data' is your DataFrame and it includes a 'species' column
fig = px.scatter(data, x="petal_length", y="petal_width", color="species",
                 labels={"petal_length": "Petal Length", "petal_width": "Petal Width"},
                 title="Petal Length vs Width (colored by species)")

fig.show()
```

```python
# plot distribution of petal length using historgam
plt.hist(data.petal_length, bins=30)
plt.xlabel('Petal Length (cm)') 
plt.ylabel('Frequency'),
plt.title('Distribution of Petal Lengths')
```

```python
# plot hostograms of numerical columns (to discover ditributions)
data.hist(bins = 20)
```

```python
# plot hostograms of numerical columns
ax = data.plot.hist(bins=25, alpha=0.5)
ax.set_xlabel('Size (cm)')
```

#### Introducing boxplots

```python
#make a box plot for each measure
data.boxplot()
```

```python
# Box plots to visualize the distribution of numerical columns
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
sns.boxplot(x='species', y='sepal_length', data=data)

plt.subplot(2, 2, 2)
sns.boxplot(x='species', y='sepal_width', data=data)

plt.subplot(2, 2, 3)
sns.boxplot(x='species', y='petal_length', data=data)

plt.subplot(2, 2, 4)
sns.boxplot(x='species', y='petal_width', data=data)
plt.show()
```

#### 7. Data Filtering and Selection
* Show how to filter data based on conditions.
* Introducing basic indexing and slicing.

```python
# Example: Selecting rows where sepal length is greater than 5.0

filtered_data = data[data['sepal_length'] > 5.0]
print("\nFiltered Data:")
print(filtered_data.head())
```

```python
filtered_data.species.value_counts()
```

#### 8. Grouping and Aggregating Data

* Grouping data by a specific column.
* Performing aggregate functions (sum, mean, count) on grouped data.

```python
# The mean of each measurement (sepal_length, sepal_width, petal_length, and petal_width).
data.groupby('species').mean()
```

```python
# The median calculation
data.groupby('species').median()
```

#### 9: Basic Statistical Analysis
* Correlation between numerical instances.

```python
# Correlation matrix
correlation_matrix = data.drop('species', axis=1).corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)
```

```python
# plot correlation matrix using seaborn
sns.heatmap(correlation_matrix, annot=True, cmap="crest")
```

```python
# Pair plot to visualize relationships between all pairs of features
sns.set_context('talk')
sns.pairplot(data, hue='species', diag_kind="hist")
```

```python
import numpy as np
import joblib
import json
```

```python
def load_class_dictionary(file_path):
    with open(file_path, 'r') as json_file:
        class_dict = json.load(json_file)
        return class_dict

def predict_price(location,sqft,bath,bhk):    
    loc_index = np.where(loaded_X.columns.str.lower()==location)[0][0]

    x = np.zeros(len(loaded_X.columns.str.lower()))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    if loc_index >= 0:
        x[loc_index] = 1

    return loaded_model.predict([x])[0]
```

```python
# Load your model and class_dict as needed
loaded_model = joblib.load('saved_model.joblib')
loaded_X = joblib.load('saved_X.joblib')
class_dict = load_class_dictionary('columns.json')
```
